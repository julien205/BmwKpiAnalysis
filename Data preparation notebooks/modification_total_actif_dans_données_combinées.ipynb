```python
import pandas as pd
```

```python
df_final = pd.read_csv('donnees_combinees.csv')
df = pd.read_csv('donnees_combinees.csv')
df.head()
```

```python
df['Type']= df['Type'].str.upper()
df = df[df['Type'] == 'GROUP']
df =df[df['Annee'].between(2016,2021)]
df
```

```python
df1 = df[df['Nom_fichier']=='Actifs_BMW']
display(df1)
df2 = df[df['Nom_fichier']=='actifs_VW']
display(df2)
df3 = df[df['Nom_fichier']=='KPI_Group_by_Annee']
display(df3)
df4 = df[df['Nom_fichier']=='Resultats_VW']
display(df4)
```

```python
df_final.loc[df3.index,'Total des actifs']=df_final.iloc[df1.index]['Total des actifs'].values
```

```python
df_final.loc[df4.index,'Total des actifs']=df_final.iloc[df2.index]['Total des actifs'].values
```

```python
df_final.to_csv('donnees_combinees.csv',index=False)
```
