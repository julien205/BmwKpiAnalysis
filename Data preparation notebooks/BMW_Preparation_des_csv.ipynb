```python
import sqlite3
import pandas as pd

# Se connecter à la base de données
conn = sqlite3.connect('ma_base_de_donnees.db')

# Récupérer la liste des tables dans la base de données
tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
table_names = [table[0] for table in tables]

# Créer un dictionnaire pour stocker les données de chaque table
data_dict = {}

# Parcourir les tables et stocker leurs données dans le dictionnaire
for table_name in table_names:
    query = f'SELECT * FROM {table_name}'
    df = pd.read_sql_query(query, conn)
    data_dict[table_name] = df

# Fusionner les données en un seul DataFrame
merged_df = pd.concat(data_dict.values(), ignore_index=True)

# Enregistrer le DataFrame fusionné dans un fichier CSV
merged_df.to_csv('donnees_combinees.csv', index=False)

# Fermer la connexion
conn.close()
```
